# **Adding a default value or filling the missing data**

```python
df.head()
```

```python
#replacing NAN with 0
df_with_0=df.fillna(0)
```

```python
df_with_0.isnull().any()
```

```python
# replacing NAN with mean
df['CALORIES'].mean()
```

```python
df_with_mean= df.fillna(df['CALORIES'].mean())
```

```python
df_with_mean.isnull().any()
```
