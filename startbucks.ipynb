```python
import pandas as pd
df = pd.read_csv('/kaggle/input/starbucks/starbucks.csv')
# Check the data types and missing values
df.info()
```

```python
# Summary statistics for numerical columns
df.describe()
```

```python
# Count of missing values in each column
df.isnull().sum()
```

```python
# Print the list of columns in the DataFrame
print(df.columns)

# Use str.contains() to find columns containing 'Total Fat'
matching_columns = df.columns[df.columns.str.contains('Total Fat')]

# Print matching columns
print(matching_columns)
```

```python
numeric_columns = [
    'Calories', 'Trans Fat (g) ', 'Saturated Fat (g)', ' Sodium (mg)',
    ' Total Carbohydrates (g) ', 'Cholesterol (mg)', ' Dietary Fibre (g)', ' Sugars (g)', ' Protein (g) '
]

# Create box plots for each numeric column
plt.figure(figsize=(12, 8))
for column in numeric_columns:
    plt.subplot(3, 3, numeric_columns.index(column) + 1)
    sns.boxplot(x=df[column])
    plt.title(f'Box Plot of {column}')

plt.tight_layout()
plt.show()
```

```python
# Visualize the distribution of numerical columns
plt.figure(figsize=(12, 6))
sns.histplot(df['Beverage_category'])  # Replace 'column_name' with an actual column name
plt.title('Distribution of Column')
plt.show()
```

```python
# # Visualize relationships between numerical columns
sns.pairplot(df[['Beverage', 'Beverage_prep', 'Calories']])  # Replace with actual column names
plt.show()
```

```python
# # Count of unique values in categorical columns
df['Beverage'].value_counts()
```

```python
# Assuming df is your DataFrame and 'Beverage' is the categorical column
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed

# Create a countplot with rotated x-axis labels
sns.countplot(x='Beverage', data=df, palette='viridis')
plt.title('Distribution of Categorical Column')
plt.xticks(rotation=35, ha='right', fontsize=10)  # Rotate and adjust font size
plt.tight_layout()  # Ensure tight layout to prevent label overlapping
plt.show()
```

```python
numeric_columns = df.select_dtypes(include='number')
# Compute the correlation matrix
correlation_matrix = numeric_columns.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Assuming df is your DataFrame and 'CoffeeColumn' is the column you want to visualize
plt.figure(figsize=(10, 6))
sns.histplot(df['Beverage'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of CoffeeColumn')
plt.xlabel('CoffeeColumn Values')
plt.ylabel('Frequency')
plt.show()
```

```python
coffee_column = df['Calories']

# Calculate skewness
skewness = coffee_column.skew()

# Visualize the distribution and mark skewness
plt.figure(figsize=(10, 6))
sns.histplot(coffee_column, bins=20, kde=True, color='skyblue')
plt.title(f'Distribution of CoffeeColumn (Skewness: {skewness:.2f})')
plt.xlabel('CoffeeColumn Values')
plt.ylabel('Frequency')
plt.show()

# Output the skewness value
print(f"Skewness of CoffeeColumn: {skewness:.2f}")
```

```python
# Assuming df is your DataFrame with the specified columns
# List of columns to consider for health evaluation
health_columns = [
    'Calories',
    'Total Carbohydrates',
    'Cholesterol',
    'sodium',
    'Dietary Fibre',
    'Sugars'
]


# Create a new column for health score (you can adjust weights based on priorities)
df['Health Score'] = (
    -0.5 * df['Calories'] +
    -0.2 * df[' Total Carbohydrates (g) '] +
    -0.3 * df['Cholesterol (mg)'] +
    -0.3 * df[' Sodium (mg)'] +
    0.4 * df[' Dietary Fibre (g)'] +
    0.4 * df[' Sugars (g)']
) 
# Identify the coffee with the highest health score
healthiest_coffee = df.loc[df['Health Score'].idxmax()]

# Display the healthiest coffee
print("The healthiest coffee is:")
print(healthiest_coffee[['Beverage_category', 'Beverage', 'Beverage_prep', 'Health Score']])
```
