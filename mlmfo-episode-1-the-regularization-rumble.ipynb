# 2. Models Face-Off 🤖

*The following information comes from "Hands-On Machine Learning with Scikit-Learn, Keras, and TensorFlow, 2nd Edition".*

The MSE(\vec \theta\right) (Mean Squared Error) formula is:
$$
\text{MSE}\left(\vec \theta\right) = \frac{1}{m}\sum_{i=1}^m(\hat{y}_i - y_i)^2
$$

Here, $m$ is the number of instances in the dataset, $\hat{y}_i$ is the predicted value by the model, and $y_i$ is the actual value. $\vec \theta$ are the weights of linear regressions, defining $\hat{y}_i$.

## What is Ridge?

> *Ridge Regression* (also called Tikhonov regularization) is a regularized version of Linear Regression: a regularization term equal to $\alpha \sum_{i=1}^n \theta_i^2$
is added to the cost function.
This forces the learning algorithm to not only fit the data but also keep the model
weights as small as possible.

Ridge cost function:

$$J\left(\vec \theta\right) = \text{MSE}\left(\vec \theta\right) + \alpha \frac{1}{2} \sum^n_{i=1}\theta_i^2$$

Here, $\alpha$ is the parameter which balances the amount of emphasis given to minimizing the residuals versus minimizing the sum of the square of the coefficients. When $\alpha = 0$, the penalty term has no effect, and Ridge Regression will produce the classical least squares estimates. However, as $\alpha$ increases, the impact of the penalty term grows, and the regression coefficients become smaller. This performs some kind of **feature scaling**.

Note that Ridge also has a classifier called "RidgeClassifier", but this notebook will focus on regression tasks.

## What is Lasso?

> *Least Absolute Shrinkage and Selection Operator Regression* (simply called *Lasso
Regression*) is another regularized version of Linear Regression: just like Ridge
Regression, it adds a regularization term to the cost function, but it uses the $\ell_1$ norm
of the weight vector instead of half the square of the $\ell_2$ norm.

Lasso Regression cost function:

$$J \left(\vec \theta \right) = \text{MSE}\left( \vec \theta \right) + \alpha \sum_{i=1}^n |\theta_i| $$

The key characteristic of Lasso Regression is that it can lead to sparse models with few coefficients; some can become exactly zero and dropped out of the model. This is particularly useful when you have a dataset with numerous features as it provides a type of **feature selection**.

```python
from sklearn.linear_model import Lasso

model = Lasso(
#_______________________________
    # Mandatory
    # There are no mandatory parameters for Lasso.

#_______________________________
    # Optional, First order

    fit_intercept=True, # Whether to calculate the intercept for this model.
    normalize=False, # This parameter is ignored when fit_intercept is set to False. If True, the regressors X will be normalized before regression by subtracting the mean and dividing by the l2-norm.
    precompute=False, # Whether to use a precomputed Gram matrix to speed up calculations.
    copy_X=True, # If True, X will be copied; else, it may be overwritten.
    max_iter=1000, # The maximum number of iterations.

#_______________________________

    warm_start=False, # When set to True, reuse the solution of the previous call to fit as initialization.
    positive=False, # When set to True, forces the coefficients to be positive.
    random_state=None, # The seed of the pseudo random number generator to use when shuffling the data.
    selection='cyclic', # If set to ‘random’, a random coefficient is updated every iteration rather than looping over features sequentially by default.
)

model.fit(X, y)
```

## A Word about Elastic Net

> *Elastic Net* is a middle ground between Ridge Regression and Lasso Regression. The
regularization term is a simple mix of both Ridge and Lasso’s regularization terms,
and you can control the mix ratio $r$. When $r = 0$, Elastic Net is equivalent to Ridge
Regression, and when $r = 1$, it is equivalent to Lasso Regression.

Elastic Net cost function:

$$J\left(\vec \theta\right) = \text{MSE}\left(\vec \theta\right) + r\alpha \sum_{i=1}^n |\theta_i| + \frac{1-r}{2}\alpha \sum^n_{i=1}\theta_i^2$$

# 3. Data Sets Preparation with Different Preprocessing 🛠️

````python
def get_encoder(df, column, max_categories=100):
    num_unique_values = df[column].nunique()
    if num_unique_values <= max_categories:
        encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)  # Set sparse=False
    else:
        encoder = OrdinalEncoder()
    return encoder



def SelectKBest_safe(k):
    while True:
        try:
            SelectKBest(k=k)
            i = True
        except: # if number of features < k
            SelectKBest_safe(k-1)
        if i == True:
            break

        

# Define three pipelines
pipeline1 = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

pipeline2 = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', MinMaxScaler()),
    ('feature_selection', SelectKBest_safe(k=10))
])

pipeline3 = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('feature_selection', RFE(estimator=Ridge(), n_features_to_select=10)) # Ridge generally used as estimator here
])


def preprocess_data(Xy, target):
    Xy = Xy.dropna(subset=[target])
    X = Xy.drop(target, axis=1)
    y = Xy[target]
    
    # Identify categorical and numerical columns
    categorical_cols = X.select_dtypes(include=['object']).columns
    numerical_cols = X.select_dtypes(exclude=['object']).columns

    # Define ColumnTransformers for each pipeline
    transformer1 = ColumnTransformer([
        ('num', pipeline1, numerical_cols),
        # Use a list comprehension to apply the appropriate encoder for each categorical column
        *[(f'cat_{col}', get_encoder(Xy, col), [col]) for col in categorical_cols]
    ])

    transformer2 = ColumnTransformer([
        ('num', pipeline2, numerical_cols),
        *[(f'cat_{col}', get_encoder(Xy, col), [col]) for col in categorical_cols]
    ])

    transformer3 = ColumnTransformer([
        ('num', pipeline3, numerical_cols),
        *[(f'cat_{col}', get_encoder(Xy, col), [col]) for col in categorical_cols]
    ])


    X1 = transformer1.fit_transform(X, y)
    X2 = transformer2.fit_transform(X, y)
    X3 = transformer3.fit_transform(X, y)
    
    return X, y, X1, X2, X3
````

# 5. First Face-Off - Execution Speed ⏱️

## Ridge is much faster than Lasso. Why?

Have a closer look at the Ridge and Lasso formula:

$$J_{Ridge}\left(\vec \theta\right) = \text{MSE}\left(\vec \theta\right) + \alpha \frac{1}{2} \sum^n_{i=1}\theta_i^2$$

$$J_{Lasso} \left(\vec \theta \right) = \text{MSE}\left( \vec \theta \right) + \alpha \sum_{i=1}^n |\theta_i| $$

Ridge adds a **quadratic** term while Lasso ads a **linear** term with a **angular point** at the origin, due to the absolute value. This angular point is critical when using gradient descent algorithm. The figure below shows the added cost function for ($z$ axis) for the $x$ and $y$ features (Ridge on the left and Lasso on the right):

![](https://www.googleapis.com/download/storage/v1/b/kaggle-forum-message-attachments/o/inbox%2F17037041%2Fed85c8ab6f176a5108ab97098822d82a%2FMLMFO%20Episode%201%20gradient.png?generation=1704491537097979&alt=media)

Have a look at the number of iterations of the gradient boosting algorithm below:

![](https://www.googleapis.com/download/storage/v1/b/kaggle-forum-message-attachments/o/inbox%2F17037041%2Fb667f185944e7f2fb48d962b0609a1d5%2FMLMFO%20Episode%201%20path%20gradient.png?generation=1704492076729674&alt=media)

You can see that because Lasso cost function has a **constant slope** on a given plane, gradient descent steps do not "descelerate" near the origin, in contrast with the Ridge curve where the slope grows linearly with distance (quadratic function, so linear slope growth). Moreover, the Lasso cost function forces the model to adopt a **smaller learning rate**, to avoid the risk to "miss" the minimum which is the angular point. With a bowl-shaped cost function, you are aware of the approach of minimum value. In Lasso, you don't know when it happened. As a result, **Lasso needs much more iterations to converge**.
