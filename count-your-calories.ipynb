```python
def convert_percent(val):
    """
    Convert the percentage string or float to an actual floating point percent
    - If the value is a string, remove %
    - Divide by 100 to make decimal
    """
    if isinstance(val, str):
        val = val.replace('%', '')
    return float(val) / 100

# Apply the conversion function and reassign the results to the DataFrame
df['Vitamin_A_DV'] = df['Vitamin_A_DV'].apply(convert_percent)
df['Vitamin_C_DV'] = df['Vitamin_C_DV'].apply(convert_percent)
df['calcium_DV'] = df['calcium_DV'].apply(convert_percent)
df['iron_DV'] = df['iron_DV'].apply(convert_percent)
```

```python
df['total_fat_g'] = df['total_fat_g'].replace('3 2', '3.2')
```

<div style="background-color: pink; padding: 10px; text-align: center; border-radius: 50px; border-style: dotted;margin-right: 180px; margin-left: 100px;">
    <h1>Data Modeling</h1>
</div>

```python
from sklearn.model_selection import train_test_split

x = df[['total_fat_g', 'trans_fat_g', 'saturated_fat_g', 'sodium_mg',
       'total_carbohydrates_g', 'cholesterol_mg', 'dietary_fibre_g', 'sugar_g',
       'protein_g', 'Vitamin_A_DV', 'Vitamin_C_DV', 'calcium_DV', 'iron_DV',
       'caffeine_mg']]
y = df['Calories']
```

```python
#splits our data into training and testing sets, with 70% of the data used for training and 30% for testing.
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression().fit(X_train, y_train)

#coefficients represent the weights assigned to each feature in the model
lm.coef_
```

```python
#coefficients alongside their corresponding feature names.
cdf = pd.DataFrame(lm.coef_, x.columns, columns=['coef'])
print(cdf)
```

<div style="background-color: pink; padding: 10px; text-align: center; border-radius: 50px; border-style: dotted;margin-right: 180px; margin-left: 100px;">
    <h1>Evaluate Model Performance</h1>
</div>

```python
#Obtain the model's predictions 
lm_predictions = lm.predict(X_test)
lm_predictions
```

```python
#Useing metrics to assess the performance of our model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print('Mean Absolute Error:',mean_absolute_error(y_test, lm_predictions)) #Lower MAE values are better
print('Mean Squared Error:',mean_squared_error(y_test, lm_predictions)) #Lower MSE values are better
print('Root Mean Squared Error:',mean_squared_error(y_test, lm_predictions, squared=False)) #Lower value is better
print('R-squared:',r2_score(y_test, lm_predictions)) #A value close to 1 indicates a very good fit
```

```python
#Visualize the relationship between predicted and actual values
sns.set_palette('RdPu_r')
sns.regplot(x=lm_predictions, y=y_test, scatter_kws={'alpha': 0.3})
plt.xlabel('predictions')
plt.title('Linear Model Evaluation')
plt.show()
```

```python
#Fixing Errors
import warnings 
warnings.filterwarnings('ignore')

#visualize the distribution of residuals
residuals = y_test-lm_predictions
sns.distplot(residuals, bins=20) 
```

```python
#check the normality assumption of the residuals
import pylab 
import scipy.stats as stats
sns.set_palette("RdPu")
fig, ax = plt.subplots()
stats.probplot(residuals, dist="norm", plot=plt, rvalue=True)
ax.get_lines()[0].set_markerfacecolor(sns.color_palette()[5])  
ax.get_lines()[1].set_color(sns.color_palette()[0])
ax.get_lines()[0].set_markersize(9.0)
ax.get_lines()[1].set_linewidth(5.0)

plt.show()
```
